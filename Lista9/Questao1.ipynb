## Seleção dos Atributos

'Time' (Tempo) representa o tempo em que a transação ocorreu com base no tempo da primeira transação da base de dados. Sua presença é irrelevante à caracterização de uma transação como fraude ou não.

```python
dados = dados.drop(columns=['Time'])
```

Com exceção de 'Time', 'Amount' e 'Class', o significado das variáveis de cada coluna é desconhecido, não podendo então serem agrupadas por seus conceitos.

### Análise de Correlação e Multicolinearidade

O agrupamento será feito no caso de se detectar alta correlação nos valores numéricos entre 2 colunas.

```python
# Matriz de correlação entre atributos
n_atributos = len(dados.columns)
atributos = dados.columns
matriz_correlacao = pd.DataFrame(index=atributos, columns=atributos)

for atributo in matriz_correlacao.columns:
    # Salvar correlação de cada atributo com cada atributo
    for atributo2 in matriz_correlacao.index:
        #print(atributo); print(atributo2)
        #correlacao = prepro.Correlation(dados[atributo], dados[atributo2])
        matriz_correlacao.loc[atributo2, atributo] = np.corrcoef(dados[atributo], dados[atributo2])[0, 1]
```

```python
matriz_correlacao
```

```python
# Para cada coluna
for coluna in matriz_correlacao.columns:
    for linha in matriz_correlacao.index:
        # Para não comparar com o próprio elemento
        if coluna != linha:
            elemento = matriz_correlacao[coluna][linha]
            if (elemento < -0.45) or (elemento > 0.45):
                print(f'[{coluna},{linha}] = {elemento}')
```

Nossos resultados indicam que a relação entre os atributos é em grande quantidade negligente (com o índice de correlação na maioria valendo 0), a única exceção sendo *V2* e *Amount* que tendem em uma relação inversa. Assim, a única coluna que podemos eliminar por correlação é V2

```python
dados = dados.drop(columns=['V2'])
```

Analisando a correlação com a classe

```python
# Para analisar a correlaçõo dos atributos com classe
correlacao_classe = []
for atributo in matriz_correlacao.index:
    if (atributo != 'Class'):
        correlacao = float(matriz_correlacao['Class'][atributo])
        # Aplicação de módulo pois o que importa é a distância de 0
        if (correlacao < 0):
            correlacao = correlacao * (-1)

        correlacao_classe.append([ atributo, correlacao ])
#print(correlacao_classe)

# Organizar na ordem decrescente
tam_correlacao = len(correlacao_classe) - 1
for i in range(tam_correlacao):
    pos_atual = tam_correlacao
    for j in range(i, tam_correlacao):
        elemento_atual = correlacao_classe[pos_atual]
        elemento_prox = correlacao_classe[pos_atual-1]
        if (elemento_atual[1] < elemento_prox[1]):
            temp = correlacao_classe[pos_atual]
            correlacao_classe[pos_atual] = correlacao_classe[pos_atual-1]
            correlacao_classe[pos_atual-1] = temp
        pos_atual -= 1
print(correlacao_classe)
```

```python
for elemento in correlacao_classe:
    if (elemento[1] < 0.01):
        print(f'{elemento[0]} tem correlação {elemento[1]} com classe')
```

Os atributos 'V13', 'V15', 'V22', 'V23', 'V24', 'V25', 'V26', 'V28' e 'Amount' apresentaram correlação negligente com o atributo de classe. Com exceção de 'Amount', os demais serão dropados.

```python
dados = dados.drop(columns=['V13', 'V15', 'V22', 'V23', 'V24', 'V25', 'V26', 'V28'])
```

## Detecção e Tratamento de Outlier

Outliers serão considerados como qualquer instância que, em um atributo qualquer, estejam a mais de 3 desvios padrões da mediana da classe. Porém, não serão considerados os dados pertencentes a classe 1 (considerado fraude), devido a baixa quantia disponível de tal.

```python
n_desviopadrao = 3

dados_copia = dados.copy()

# Atributos a serem analisados
colunas = dados_copia.columns
colunas = colunas.drop(['Class'])

#print(dados.iloc[6672])
for coluna in colunas:
    z_scores = np.abs( stats.zscore(dados_copia[coluna]) )
    outliers_indices = np.where( z_scores > n_desviopadrao )[0]
    print(f'Coluna {coluna}, outliers detectados = {len(outliers_indices)}')
    to_drop = []
    for instancia in outliers_indices:
        #print(f'instancia: {instancia} ({dados_copia.iloc[instancia].name})')
        #print(dados.iloc[instancia]); print("--- --- ---")

        #dados_copia = dados_copia.drop(index=dados_copia.iloc[instancia].name)
        #dados_copia = dados_copia.drop(index=instancia)

        #to_drop.append(dados_copia.iloc[instancia].name)
        # Para evitar dropar as instâncias de fraude
        if (dados_copia['Class'][dados_copia.iloc[instancia].name] != 1 ):
            to_drop.append(dados_copia.iloc[instancia].name)

    dados_copia = dados_copia.drop(index=to_drop)
dados = dados_copia.copy()
```

## Balanceamento da Classe

Serão comparados resultados sem balanceamento, com undersampling, com oversampling e com tanto undersampling quanto oversampling

### Undersampling

```python
balanceamento_under = TomekLinks(sampling_strategy='auto')

# Undersampling das instâncias de treino
atributos_treino_under, classe_treino_under = balanceamento_under.fit_resample(atributos_treino, classe_treino)
```

### Oversampling

```python
balanceamento_over = SMOTE(sampling_strategy='minority')

# Oversampling das instâncias de treino
atributos_treino_over, classe_treino_over = balanceamento_over.fit_resample(atributos_treino, classe_treino)
atributos_treino_underover, classe_treino_underover = balanceamento_over.fit_resample(atributos_treino_under, classe_treino_under)
```

```python
# Contar número de resultados no atributo de classe e quais são
resultados = list(set(dados['Class']))
n_classe = len(resultados)
matriz_comparacao = pd.DataFrame(index=resultados, columns=['Não Pré-Processado', 'Não Balanceado', 'Under', 'Over', 'Under/Over'])

# Inicializar valores em 0
for coluna in matriz_comparacao.columns:
    for linha in matriz_comparacao.index:
        matriz_comparacao.loc[linha, coluna] = 0
```

# Modelo Preditivo

Por conter dados numéricos, foi escolhido o uso de redes neurais.

O número de neurônios na camada intermediaria foi definido usando a regra do valor médio ((p+M)/2)

```python
# Parâmetros
entradas = len(dados.columns)
saidas = 1
n_intermediario = math.ceil((entradas+saidas)/2)

activation = 'relu'

learning_rate = 'adaptive'
learning_rate_initial = 0.175

max_iterations = 7000

error_tolerance = 0.05
n_iterations_to_change = 20

rede_neural_nprocessada = MLPClassifier(hidden_layer_sizes=n_intermediario, activation=activation, learning_rate=learning_rate, learning_rate_init=learning_rate_initial, max_iter=max_iterations, random_state=69, tol=error_tolerance, n_iter_no_change=n_iterations_to_change)

rede_neural_nbalanceada = MLPClassifier(hidden_layer_sizes=n_intermediario, activation=activation, learning_rate=learning_rate, learning_rate_init=learning_rate_initial, max_iter=max_iterations, random_state=69, tol=error_tolerance, n_iter_no_change=n_iterations_to_change)
rede_neural_under = MLPClassifier(hidden_layer_sizes=n_intermediario, activation=activation, learning_rate=learning_rate, learning_rate_init=learning_rate_initial, max_iter=max_iterations, random_state=69, tol=error_tolerance, n_iter_no_change=n_iterations_to_change)
rede_neural_over = MLPClassifier(hidden_layer_sizes=n_intermediario, activation=activation, learning_rate=learning_rate, learning_rate_init=learning_rate_initial, max_iter=max_iterations, random_state=69, tol=error_tolerance, n_iter_no_change=n_iterations_to_change)
rede_neural_underover = MLPClassifier(hidden_layer_sizes=n_intermediario, activation=activation, learning_rate=learning_rate, learning_rate_init=learning_rate_initial, max_iter=max_iterations, random_state=69, tol=error_tolerance, n_iter_no_change=n_iterations_to_change)
```

```python
rede_neural_nprocessada = rede_neural_nprocessada.fit(X= atributos_treino_original, y= classe_treino_original)
rede_neural_nbalanceada = rede_neural_nbalanceada.fit(X= atributos_treino, y= classe_treino)
rede_neural_under = rede_neural_under.fit(X= atributos_treino_under, y= classe_treino_under)
rede_neural_over = rede_neural_over.fit(X= atributos_treino_over, y= classe_treino_over)
rede_neural_underover = rede_neural_underover.fit(X= atributos_treino_underover, y= classe_treino_underover)
```
